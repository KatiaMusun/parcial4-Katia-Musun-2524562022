Preparar los datos en formato adecuado para aplicar reglas de asociación.

```python

basket = (
    df.groupby(['transaccion_id', 'item'])['item']
    .count()
    .unstack()
    .fillna(0)
)


basket = basket.applymap(lambda x: 1 if x > 0 else 0)

print("\nMatriz transaccional:")
display(basket.head())
```

Aplicar el algoritmo Apriori o una técnica equivalente.

```python

frequent_items = apriori(
    basket,
    min_support=0.05,
    use_colnames=True
)

print("\nItemsets frecuentes:")
display(frequent_items.head())
```

Generar reglas de asociación usando soporte, confianza y lift.

```python
rules = association_rules(
    frequent_items,
    metric="confidence",
    min_threshold=0.4
)

# Ordenar por lift
rules = rules.sort_values(by='lift', ascending=False)
```

Mostrar las 10 reglas más relevantes.

```python
top_rules = rules[[
    'antecedents',
    'consequents',
    'support',
    'confidence',
    'lift'
]].head(10)

print("\nTop 10 reglas de asociación:")
display(top_rules)
```
